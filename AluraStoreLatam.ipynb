# 3. Calificación promedio de la tienda


```python
# calcular las calificaciones promedio de los clientes para cada tienda

import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# TIENDA 1
tienda['Calificación'] = pd.to_numeric(tienda['Calificación'], errors='coerce')
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio de Tienda 1: {calificacion_promedio_tienda1}")

# TIENDA 2
tienda2['Calificación'] = pd.to_numeric(tienda2['Calificación'], errors='coerce')
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio de Tienda 2: {calificacion_promedio_tienda2}")

# TIENDA 3
tienda3['Calificación'] = pd.to_numeric(tienda3['Calificación'], errors='coerce')
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio de Tienda 3: {calificacion_promedio_tienda3}")

# TIENDA 4
tienda4['Calificación'] = pd.to_numeric(tienda4['Calificación'], errors='coerce')
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio de Tienda 4: {calificacion_promedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python
#qué productos fueron los más vendidos y los menos vendidos en cada tienda

# TIENDA 1
productos_vendidos_tienda1 = tienda.groupby('Producto').size()
producto_mas_vendido_tienda1 = productos_vendidos_tienda1.idxmax()
producto_menos_vendido_tienda1 = productos_vendidos_tienda1.idxmin()
print(f"Tienda 1 - Producto más vendido: {producto_mas_vendido_tienda1}")
print(f"Tienda 1 - Producto menos vendido: {producto_menos_vendido_tienda1}")

# TIENDA 2
productos_vendidos_tienda2 = tienda2.groupby('Producto').size()
producto_mas_vendido_tienda2 = productos_vendidos_tienda2.idxmax()
producto_menos_vendido_tienda2 = productos_vendidos_tienda2.idxmin()
print(f"Tienda 2 - Producto más vendido: {producto_mas_vendido_tienda2}")
print(f"Tienda 2 - Producto menos vendido: {producto_menos_vendido_tienda2}")

# TIENDA 3
productos_vendidos_tienda3 = tienda3.groupby('Producto').size()
producto_mas_vendido_tienda3 = productos_vendidos_tienda3.idxmax()
producto_menos_vendido_tienda3 = productos_vendidos_tienda3.idxmin()
print(f"Tienda 3 - Producto más vendido: {producto_mas_vendido_tienda3}")
print(f"Tienda 3 - Producto menos vendido: {producto_menos_vendido_tienda3}")

# TIENDA 4
productos_vendidos_tienda4 = tienda4.groupby('Producto').size()
producto_mas_vendido_tienda4 = productos_vendidos_tienda4.idxmax()
producto_menos_vendido_tienda4 = productos_vendidos_tienda4.idxmin()
print(f"Tienda 4 - Producto más vendido: {producto_mas_vendido_tienda4}")
print(f"Tienda 4 - Producto menos vendido: {producto_menos_vendido_tienda4}")
```

```python
# Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda

import matplotlib.pyplot as plt
import seaborn as sns

# Helper function to visualize top N products by sales for a given store DataFrame
def visualize_top_products(df, store_name, n=5):
  # Group by product and sum the 'Precio' to get total sales per product
  product_sales = df.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

  # Get the top N products
  top_products = product_sales.head(n)

  plt.figure(figsize=(10, 6))
  sns.barplot(x=top_products.index, y=top_products.values)
  plt.title(f'Top {n} Productos por Ventas en {store_name}')
  plt.xlabel('Producto')
  plt.ylabel('Ventas Totales')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()

# Visualize top products for each store
visualize_top_products(tienda, 'Tienda 1')
visualize_top_products(tienda2, 'Tienda 2')
visualize_top_products(tienda3, 'Tienda 3')
visualize_top_products(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
# calcular el costo de envío promedio para cada tienda

import pandas as pd
# TIENDA 1
tienda['Costo de envío'] = pd.to_numeric(tienda['Costo de envío'], errors='coerce')
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
print(f"Costo de envío promedio de Tienda 1: {costo_envio_promedio_tienda1}")

# TIENDA 2
tienda2['Costo de envío'] = pd.to_numeric(tienda2['Costo de envío'], errors='coerce')
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
print(f"Costo de envío promedio de Tienda 2: {costo_envio_promedio_tienda2}")

# TIENDA 3
tienda3['Costo de envío'] = pd.to_numeric(tienda3['Costo de envío'], errors='coerce')
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
print(f"Costo de envío promedio de Tienda 3: {costo_envio_promedio_tienda3}")

# TIENDA 4
tienda4['Costo de envío'] = pd.to_numeric(tienda4['Costo de envío'], errors='coerce')
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
print(f"Costo de envío promedio de Tienda 4: {costo_envio_promedio_tienda4}")
```

```python
# cuánto se gasta, en promedio, en el envío de cada tienda

import matplotlib.pyplot as plt
# Calculate the average shipping cost for each store
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Gasto promedio en envío de Tienda 1: {costo_envio_promedio_tienda1}")
print(f"Gasto promedio en envío de Tienda 2: {costo_envio_promedio_tienda2}")
print(f"Gasto promedio en envío de Tienda 3: {costo_envio_promedio_tienda3}")
print(f"Gasto promedio en envío de Tienda 4: {costo_envio_promedio_tienda4}")

# Create a bar chart to visualize the average shipping costs
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
avg_shipping_costs = [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]

plt.figure(figsize=(8, 6))
sns.barplot(x=stores, y=avg_shipping_costs)
plt.title('Gasto Promedio en Envío por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.show()
```
