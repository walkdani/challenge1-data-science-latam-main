# 2. Ventas por categoría

```python
#cuantas ventas se hicieron por categoria teniendo en cuenta cada una de las tiendas

ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 1:")
print(ventas_por_categoria_tienda1)

ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 2:")
print(ventas_por_categoria_tienda2)

ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 3:")
print(ventas_por_categoria_tienda3)

ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 4:")
ventas_por_categoria_tienda4
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 1:")
print(ventas_por_categoria_tienda1)

ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 2:")
print(ventas_por_categoria_tienda2)

ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 3:")
print(ventas_por_categoria_tienda3)

ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print("\nVentas por categoría en Tienda 4:")
ventas_por_categoria_tienda4

all_stores = pd.concat([tienda, tienda2, tienda3, tienda4])
ventas_por_categoria_total = all_stores.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=ventas_por_categoria_total.index, y=ventas_por_categoria_total.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Ventas Totales por Categoría en Todas las Tiendas')
plt.xlabel('Categoría del Producto')
plt.ylabel('Total de Ventas')
plt.tight_layout()
plt.show()

# Plotting individual store sales by category
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Ventas por Categoría por Tienda')

# Tienda 1
sns.barplot(x=ventas_por_categoria_tienda1.index, y=ventas_por_categoria_tienda1.values, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_xticklabels(ventas_por_categoria_tienda1.index, rotation=45, ha='right')
axes[0, 0].set_title('Tienda 1')
axes[0, 0].set_xlabel('Categoría del Producto')
axes[0, 0].set_ylabel('Ventas')

# Tienda 2
sns.barplot(x=ventas_por_categoria_tienda2.index, y=ventas_por_categoria_tienda2.values, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_xticklabels(ventas_por_categoria_tienda2.index, rotation=45, ha='right')
axes[0, 1].set_title('Tienda 2')
axes[0, 1].set_xlabel('Categoría del Producto')
axes[0, 1].set_ylabel('Ventas')

# Tienda 3
sns.barplot(x=ventas_por_categoria_tienda3.index, y=ventas_por_categoria_tienda3.values, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_xticklabels(ventas_por_categoria_tienda3.index, rotation=45, ha='right')
axes[1, 0].set_title('Tienda 3')
axes[1, 0].set_xlabel('Categoría del Producto')
axes[1, 0].set_ylabel('Ventas')

# Tienda 4
sns.barplot(x=ventas_por_categoria_tienda4.index, y=ventas_por_categoria_tienda4.values, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_xticklabels(ventas_por_categoria_tienda4.index, rotation=45, ha='right')
axes[1, 1].set_title('Tienda 4')
axes[1, 1].set_xlabel('Categoría del Producto')
axes[1, 1].set_ylabel('Ventas')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# calcular las calificaciones promedio de los clientes para cada tienda

import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# TIENDA 1
tienda['Calificación'] = pd.to_numeric(tienda['Calificación'], errors='coerce')
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio de Tienda 1: {calificacion_promedio_tienda1}")

# TIENDA 2
tienda2['Calificación'] = pd.to_numeric(tienda2['Calificación'], errors='coerce')
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio de Tienda 2: {calificacion_promedio_tienda2}")

# TIENDA 3
tienda3['Calificación'] = pd.to_numeric(tienda3['Calificación'], errors='coerce')
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio de Tienda 3: {calificacion_promedio_tienda3}")

# TIENDA 4
tienda4['Calificación'] = pd.to_numeric(tienda4['Calificación'], errors='coerce')
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio de Tienda 4: {calificacion_promedio_tienda4}")
```

```python
# generar y visualizar un grafico de pastel con respecto al ejercicio calificacion promedio tiendas

import matplotlib.pyplot as plt
import pandas as pd
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# TIENDA 1
tienda['Calificación'] = pd.to_numeric(tienda['Calificación'], errors='coerce')
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio de Tienda 1: {calificacion_promedio_tienda1}")

# TIENDA 2
tienda2['Calificación'] = pd.to_numeric(tienda2['Calificación'], errors='coerce')
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio de Tienda 2: {calificacion_promedio_tienda2}")

# TIENDA 3
tienda3['Calificación'] = pd.to_numeric(tienda3['Calificación'], errors='coerce')
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio de Tienda 3: {calificacion_promedio_tienda3}")

# TIENDA 4
tienda4['Calificación'] = pd.to_numeric(tienda4['Calificación'], errors='coerce')
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio de Tienda 4: {calificacion_promedio_tienda4}")

# Datos para el grafico circular
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
sizes = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Calificación Promedio por Tienda')
plt.show()
```
